```python
import os

# Klas√∂rler
os.makedirs("app/llm_engine", exist_ok=True)
os.makedirs("app/storage", exist_ok=True)

# __init__.py dosyalarƒ±
open("app/__init__.py", "w").close()
open("app/llm_engine/__init__.py", "w").close()
open("app/storage/__init__.py", "w").close()

# --- LLM Motoru (Streaming) ---
generator_code = r'''
import os
from openai import OpenAI

def generate_story(theme: str, age: int, story_type: str, length_key: str):
    api_key = os.getenv("OPENAI_API_KEY", "")
    client = OpenAI(api_key=api_key)

    length_map = {
        "short":  "200-300 kelime",
        "medium": "500-700 kelime",
        "long":   "1000+ kelime, mutlaka final yap"
    }

    system_prompt = (
        "Sen usta bir masal yazarƒ±sƒ±n. T√ºrk√ße yaz. K√ºf√ºr/nefret/≈üiddeti √∂zendirme yok. "
        f"Hedef ya≈ü: {age}. T√ºr: {story_type}. "
        "Giri≈ü-Geli≈üme-Sonu√ß yapƒ±sƒ± kur. Tutarlƒ± bir final yap."
    )

    user_prompt = (
        f"Konu: {theme}\n"
        f"Uzunluk: {length_map.get(length_key,'medium')}\n"
        "√áƒ±ktƒ± bi√ßimi:\n"
        "1) Ba≈ülƒ±k\n"
        "2) Masal metni\n"
        "3) En sonda 3 maddelik 'Masaldan ders' b√∂l√ºm√º"
    )

    # Hata y√∂netimi ekleyelim
    try:
        stream = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role":"system","content":system_prompt},
                {"role":"user","content":user_prompt},
            ],
            stream=True,
            max_tokens=1800
        )

        for chunk in stream:
            delta = chunk.choices[0].delta
            if delta and getattr(delta, "content", None):
                yield delta.content
    except Exception as e:
        yield f"HATA: API baƒülantƒ±sƒ±nda sorun olu≈ütu. L√ºtfen API Key'inizi kontrol edin. Hata detayƒ±: {e}"
'''
with open("app/llm_engine/generator.py", "w", encoding="utf-8") as f:
    f.write(generator_code)

# --- Kayƒ±t mod√ºl√º (JSON) ---
storage_code = r'''
import json, os, time, uuid

DB_PATH = os.path.join("app","storage","stories.json")

def _load():
    if not os.path.exists(DB_PATH):
        return []
    try:
        with open(DB_PATH,"r",encoding="utf-8") as f:
            return json.load(f)
    except Exception:
        return []

def _save(items):
    with open(DB_PATH,"w",encoding="utf-8") as f:
        json.dump(items,f,ensure_ascii=False,indent=2)

def add_story(title, theme, story_type, age, length, content):
    items = _load()
    item = {
        "id": str(uuid.uuid4()),
        "created_at": int(time.time()),
        "title": title,
        "theme": theme,
        "story_type": story_type,
        "age": age,
        "length": length,
        "content": content
    }
    items.insert(0, item)
    _save(items)
    return item["id"]

def list_stories():
    return _load()

def get_story(story_id):
    for it in _load():
        if it.get("id") == story_id:
            return it
    return None

def delete_story(story_id):
    items = _load()
    new_items = [it for it in items if it.get("id") != story_id]
    _save(new_items)
    return len(new_items) != len(items)
'''
with open("app/storage/store.py", "w", encoding="utf-8") as f:
    f.write(storage_code)

# --- PDF export (T√úRK√áE KARAKTER D√úZELTMELƒ∞) ---
# FPDF font y√ºklemeden T√ºrk√ße karakter basamaz. Harfler kaybolmasƒ±n diye
# basit bir harf deƒüi≈ütirme (mapping) fonksiyonu ekledim.
pdf_export_code = r'''
from fpdf import FPDF

def safe_text(text: str) -> str:
    if text is None: return ""
    text = str(text)
    # T√ºrk√ße karakterleri en yakƒ±n Latin karakterine √ßevir (Harf yutmasƒ±n diye)
    replacements = {
        "ƒü": "g", "ƒû": "G", "√º": "u", "√ú": "U", "≈ü": "s", "≈û": "S",
        "ƒ±": "i", "ƒ∞": "I", "√∂": "o", "√ñ": "O", "√ß": "c", "√á": "C"
    }
    for tr, eng in replacements.items():
        text = text.replace(tr, eng)

    return text.encode("latin-1", "ignore").decode("latin-1")

def story_to_pdf_bytes(title: str, content: str) -> bytes:
    pdf = FPDF()
    pdf.add_page()
    pdf.set_auto_page_break(auto=True, margin=15)

    pdf.set_font("Arial", "B", 14)
    pdf.multi_cell(0, 8, safe_text(title))
    pdf.ln(2)

    pdf.set_font("Arial", "", 11)
    pdf.multi_cell(0, 6, safe_text(content))

    out = pdf.output(dest="S")
    if isinstance(out, str):
        return out.encode("latin-1", "ignore")
    return bytes(out)
'''
with open("app/storage/pdf_export.py", "w", encoding="utf-8") as f:
    f.write(pdf_export_code)

print("‚úÖ Dosyalar g√ºncellendi (app/llm_engine ve app/storage). ≈ûimdi main.py kodunu √ßalƒ±≈ütƒ±rabilirsin.")
```

```python
import streamlit as st
import re
import time

# -----------------------------
# Gerekli Mod√ºller
# -----------------------------
try:
    from app.llm_engine.generator import generate_story
    from app.storage.store import add_story, list_stories, get_story, delete_story
    from app.storage.pdf_export import story_to_pdf_bytes
except ImportError:
    st.error("L√ºtfen √∂nce kurulum dosyalarƒ±nƒ± √ßalƒ±≈ütƒ±rdƒ±ƒüƒ±nƒ±zdan emin olun.")
    st.stop()

# -----------------------------
# Sayfa Konfig√ºrasyonu
# -----------------------------
st.set_page_config(
    page_title="Masal At√∂lyesi PRO",
    page_icon="üìö",
    layout="wide",
    initial_sidebar_state="collapsed"
)

# -----------------------------
# Session State
# -----------------------------
if "last_story_text" not in st.session_state: st.session_state.last_story_text = ""
if "last_story_title" not in st.session_state: st.session_state.last_story_title = ""
if "last_story_meta" not in st.session_state: st.session_state.last_story_meta = {}
if "saved_toast" not in st.session_state: st.session_state.saved_toast = False
if "open_story_id" not in st.session_state: st.session_state.open_story_id = None
if "force_refresh" not in st.session_state: st.session_state.force_refresh = 0

# -----------------------------
# CSS TASARIM (KOYU TEMA + KARTLAR)
# -----------------------------
st.markdown(
    """
<style>
    header[data-testid="stHeader"] { visibility: hidden; }
    div[data-testid="stToolbar"] { visibility: hidden; height: 0; }
    .block-container { padding-top: 2rem !important; }

    /* GENEL */
    .stApp { background-color: #0E1117 !important; }
    h1, h2, h3, h4, p, span, div, label { color: #FFFFFF !important; }

    /* KART TASARIMI */
    .custom-card {
        background-color: #1E2329 !important;
        border: 1px solid #333 !important;
        border-radius: 12px;
        padding: 20px;
        margin-bottom: 20px;
        box-shadow: 0 4px 6px rgba(0,0,0,0.3);
    }

    .custom-card h3 { color: #FF4B4B !important; margin-top:0; }
    .meta-text { color: #AAAAAA !important; font-size: 0.9rem; margin-bottom: 10px; }
    .content-preview { color: #DDD !important; font-size: 0.95rem; }

    /* Input Alanlarƒ± */
    div[data-testid="stTextInput"] input, div[data-testid="stSelectbox"] div[role="combobox"] {
        background-color: #262730 !important;
        color: white !important;
        border: 1px solid #444 !important;
    }

    /* Butonlar */
    div.stButton > button {
        width: 100%;
        border-radius: 8px;
        font-weight: bold;
        background-color: #FF4B4B;
        color: white;
        border: none;
    }
    div.stButton > button:hover { background-color: #FF2B2B; }

    /* Silme butonu */
    .trash-btn button {
        background-color: #333 !important;
        color: #ff6b6b !important;
    }
</style>
    """,
    unsafe_allow_html=True,
)

# -----------------------------
# BA≈ûLIK
# -----------------------------
st.markdown("""
<div style="text-align:center; padding: 20px 0; margin-bottom: 20px;">
    <h1 style="color:white; margin:0;">üìö Masal At√∂lyesi PRO</h1>
    <p style="color:#888; margin:5px 0 0 0;">Yapay Zeka Destekli Hikaye Olu≈üturucu</p>
</div>
""", unsafe_allow_html=True)

tab1, tab2 = st.tabs(["‚ú® YENƒ∞ MASAL", "üìÇ K√úT√úPHANE & ARAMA"])

# -----------------------------
# Yardƒ±mcƒ± Fonksiyon
# -----------------------------
def parse_story(text):
    text = (text or "").strip()
    lines = [ln.strip() for ln in text.splitlines() if ln.strip()]
    title = lines[0] if lines else "Adsƒ±z Masal"
    title = re.sub(r"^#+\s*", "", title).replace("Ba≈ülƒ±k:", "").strip()

    lower = text.lower()
    idx = lower.find("masaldan ders")
    content = text[:idx].strip() if idx != -1 else text
    lessons = []

    if idx != -1:
        part = text[idx:].strip()
        for ln in part.splitlines():
            clean = ln.strip()
            if re.match(r"^(\d+[\).\]]|\-|\‚Ä¢)\s+", clean):
                lessons.append(re.sub(r"^(\d+[\).\]]|\-|\‚Ä¢)\s+", "", clean).strip())

    return title, content, lessons

# -----------------------------
# SEKME 1: √úRETƒ∞M
# -----------------------------
with tab1:
    c1, c2 = st.columns([1, 2], gap="medium")

    with c1:
        st.markdown("### ‚öôÔ∏è Ayarlar")
        theme = st.text_input("Konu", placeholder="√ñrn: Uzayda piknik")
        genre = st.selectbox("T√ºr", ["Macera", "Fantastik", "Bilim Kurgu", "Komedi", "Duygusal", "Eƒüitici"])
        age = st.slider("Ya≈ü", 3, 15, 7)
        length_ui = st.select_slider("Uzunluk", ["Kƒ±sa", "Orta", "Uzun"], value="Kƒ±sa")
        length_key = {"Kƒ±sa":"short", "Orta":"medium", "Uzun":"long"}[length_ui]

        st.markdown("---")
        if st.button("üöÄ Masalƒ± Yaz", type="primary"):
            if not theme:
                st.warning("L√ºtfen bir konu yazƒ±n.")
            else:
                with c2:
                    ph = st.empty()
                    full = ""
                    for chunk in generate_story(theme, age, genre, length_key):
                        full += chunk
                        ph.markdown(f'<div class="custom-card">{full}</div>', unsafe_allow_html=True)

                    st.session_state.last_story_text = full
                    t, c, l = parse_story(full)
                    st.session_state.last_story_title = t
                    st.session_state.last_story_meta = {"theme":theme, "genre":genre, "age":age, "len":length_ui}
                    st.rerun()

    with c2:
        if st.session_state.saved_toast:
            st.success("‚úÖ Kaydedildi!")
            st.session_state.saved_toast = False

        if st.session_state.last_story_text:
            title, content, lessons = parse_story(st.session_state.last_story_text)
            meta = st.session_state.last_story_meta

            st.markdown(f"""
            <div class="custom-card">
                <h3>{title}</h3>
                <div class="meta-text">{meta.get('genre')} ‚Ä¢ {meta.get('age')} Ya≈ü ‚Ä¢ {meta.get('len')}</div>
                <div style="white-space: pre-wrap; line-height:1.6;">{content}</div>
                {'<br><b>‚ú® Dersler:</b><br>' + '<br>'.join([f'- {x}' for x in lessons]) if lessons else ''}
            </div>
            """, unsafe_allow_html=True)

            b1, b2 = st.columns(2)
            with b1:
                if st.button("üíæ K√ºt√ºphaneye Kaydet"):
                    add_story(title, meta['theme'], meta['genre'], meta['age'], "short", st.session_state.last_story_text)
                    st.session_state.saved_toast = True
                    st.rerun()
            with b2:
                try:
                    data = story_to_pdf_bytes(title, st.session_state.last_story_text)
                    st.download_button("üìÑ PDF ƒ∞ndir", data=data, file_name="masal.pdf", mime="application/pdf")
                except: pass

# -----------------------------
# SEKME 2: K√úT√úPHANE & ARAMA
# -----------------------------
with tab2:
    # Arama placeholder'ƒ±nƒ± g√ºncelledik
    search = st.text_input("üîç K√ºt√ºphanede Ara (Ba≈ülƒ±k, Konu veya T√ºr...)", placeholder="√ñrn: Macera, kedi, uzay...")

    items = list_stories()

    # --- GELƒ∞≈ûMƒ∞≈û ARAMA MANTIƒûI BURADA ---
    if search:
        q = search.lower()
        filtered_items = []
        for x in items:
            # Ba≈ülƒ±k + Konu + T√ºr alanlarƒ±nƒ± birle≈ütirip arƒ±yoruz
            combined_text = (
                str(x.get('title', '')) + " " +
                str(x.get('theme', '')) + " " +
                str(x.get('story_type', ''))
            ).lower()

            if q in combined_text:
                filtered_items.append(x)
        items = filtered_items
    # -------------------------------------

    if not items:
        if search:
            st.info(f"'{search}' ile ilgili sonu√ß bulunamadƒ±.")
        else:
            st.info("Kayƒ±tlƒ± masal yok.")
    else:
        st.caption(f"{len(items)} masal listeleniyor")
        for it in items:
            st.markdown(f"""
            <div class="custom-card" style="padding:15px; margin-bottom:10px;">
                <h3 style="margin:0;">{it.get('title','Adsƒ±z')}</h3>
                <div class="meta-text">
                    <span style="background:#333; padding:2px 6px; border-radius:4px; color:#fff;">{it.get('story_type')}</span>
                    ‚Ä¢ {it.get('age')} Ya≈ü ‚Ä¢ Konu: {it.get('theme')}
                </div>
                <div class="content-preview">{it.get('content','').replace(chr(10),' ')[:130]}...</div>
            </div>
            """, unsafe_allow_html=True)

            c_a, c_b = st.columns([1, 4])
            with c_a:
                if st.button("üìñ Oku", key=f"r_{it['id']}"):
                    st.session_state.open_story_id = it['id']
                    st.rerun()
            with c_b:
                st.markdown('<div class="trash-btn">', unsafe_allow_html=True)
                if st.button("üóëÔ∏è Sil", key=f"d_{it['id']}"):
                    delete_story(it['id'])
                    if st.session_state.open_story_id == it['id']: st.session_state.open_story_id = None
                    st.rerun()
                st.markdown('</div>', unsafe_allow_html=True)

    if st.session_state.open_story_id:
        st.markdown("---")
        s = get_story(st.session_state.open_story_id)
        if s:
            t, c, l = parse_story(s.get("content",""))
            st.markdown(f"""
            <div class="custom-card">
                <h2 style="text-align:center; color:#FF4B4B !important;">{t}</h2>
                <div style="white-space: pre-wrap; font-size:1.1rem;">{c}</div>
                {'<div style="margin-top:20px; border-top:1px solid #444; padding-top:10px;"><b>üåü Dersler:</b><br>' + '<br>'.join([f'- {x}' for x in l]) + '</div>' if l else ''}
            </div>
            """, unsafe_allow_html=True)
            data = story_to_pdf_bytes(t, s.get("content",""))
            st.download_button("PDF Olarak Al", data=data, file_name="arsiv.pdf", mime="application/pdf")
```
